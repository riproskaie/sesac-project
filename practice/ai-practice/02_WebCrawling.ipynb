#### 문제1: 링크된 페이지에서 크롬 개발자 도구를 이용해 로딩할 때 데이터를 어떻게 가져오는지 확인해 보고, Request를 모방해서 응답받은 JSON을 딕셔너리로 이루어진 리스트를 `top_movies` 리스트에 저장하세요. 

```python
import requests

```

```python
import time
import random
top_movies = []
for i in range(35):
    response = requests.get(f'https://pedia.watcha.com/api/staffmades/278/contents?page={i}&size=9', cookies=cookies, headers=headers)
    response = response.json()['result']['result']
    top_movies.extend(response)
    time.sleep(random.randrange(1,5))
    
```

#### 문제 2: 수집한 `top_movies` 리스트 안에 있는 딕셔너리에서 `code`, `title`, `year`, `ratings_avg`, `nations`, `genres`만 남기고 나머지는 필요 없으니 없애버리세요.

```python
new_top_movies = []
for tm in top_movies:
    new_top_movies.extend(tm)
```

```python
top_movies = new_top_movies
```

```python
cleaned_top_movies = []
for movie in top_movies:
    code = movie['code']
    title = movie['title']
    year = movie['year']
    ratings_avg = movie['ratings_avg']
    nations = [ m['name'] for m in movie['nations']]
    genres = movie['genres']
    movie_dict = {
        "code": code,
        "title": title,
        "year": year,
        "ratings_avg" : ratings_avg,
        "nations": nations,
        "genres": genres
    }
    cleaned_top_movies.append(movie_dict)
```

```python
cleaned_top_movies
```

#### 문제3: `cleaned_top_movies`를 데이터 프레임으로 만들고, `watcha_top_movies.csv` 파일로 저장하세요.

```python
top_movies_df = pd.DataFrame(cleaned_top_movies)
top_movies_df
```

```python
top_movies_df = top_movies_df.drop_duplicates(['code'])
top_movies_df = top_movies_df.reset_index().iloc[:, 1:]
top_movies_df.to_csv("watcha_top_movies.csv",index=False)
```
